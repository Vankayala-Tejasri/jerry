```python
from sklearn.tree import DecisionTreeClassifier
decision=DecisionTreeClassifier()
```

```python
# prompt: now take the above dataset print total columns      list

print("Total columns in the dataset:", df.columns.tolist())
```

```python
# prompt: now construct the decision tree classifer based on the dataset i given

X = df.iloc[:, :-1]
y = df.iloc[:, -1]
decision.fit(X, y)
```

```python
# prompt: now do all the process after fitting
from sklearn.metrics import classification_report
# Predict the values for the test set
y_pred = decision.predict(X)

# Calculate the accuracy of the model
accuracy = decision.score(X, y)
print("Accuracy:", accuracy)

# Calculate the confusion matrix
# confusion_matrix = pd.crosstab(y, y_pred, rownames=['Actual'], colnames=['Predicted'])
# print("Confusion Matrix:\n", confusion_matrix)

# Calculate the classification report
classification_report = pd.DataFrame(classification_report(y, y_pred, output_dict=True))
print("Classification Report:\n", classification_report)

# Visualize the decision tree
from sklearn.tree import plot_tree
plt.figure(figsize=(10, 10))
plot_tree(decision, filled=True)
plt.show()
```

```python
# prompt: now i have the test dataset check the decision tree with the test dataset
from sklearn.metrics import classification_report
# Load the test dataset
test_data = pd.read_csv('/content/drive/MyDrive/Testing.csv')

# Remove any duplicate rows from the test dataset
test_data = test_data.drop_duplicates()

# Extract the features and labels from the test dataset
X_test = test_data.iloc[:, :-1]
y_test = test_data.iloc[:, -1]

# Predict the labels for the test dataset using the trained decision tree
y_pred_test = decision.predict(X_test)

# Calculate the accuracy of the model on the test dataset
accuracy_test = decision.score(X_test, y_test)
print("Accuracy on the test dataset:", accuracy_test)

# Calculate the confusion matrix for the test dataset
# confusion_matrix_test = pd.crosstab(y_test, y_pred_test, rownames=['Actual'], colnames=['Predicted'])
# print("Confusion Matrix on the test dataset:\n", confusion_matrix_test)
classification_report_test = classification_report(y_test, y_pred_test, output_dict=True)
print("Classification Report on the test dataset:\n", classification_report_test)
```
